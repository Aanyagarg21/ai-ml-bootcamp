# SQL [Structured Query Language]
- Standardize programming language
- Used to:
   - Manage and Manipulate Relational databases
   - Perform Operations like: Querying,Updating and ...


## SQL Types:
- MySQL : open-source relational db management system
- PostgreSQL : advanced, open-source object-relational db
- SQLite : lightweight, serverless SQL db, used for testing and prototyping
- Microsoft SQL server :
- Oracle : multi-model db management system

```python
import numpy as np
import pandas as pd
import sqlite3 as sql
```

✨Define Dataset Parameters ✨
</div>

```python
np.random.seed(20)
n_student = 100
first_names = ["Emma", "Noah", "Olivia", "Liam", "Ava", "Ethan", "Sophia", "Mason", "Isabella", "William"]
cities = ["Delhi", "Mumbai", "Bangalore", "Hyderabad", "Chennai", "Kolkata", "Pune", "Ahmedabad", "Jaipur", "Lucknow"]
```

<div style="

">
✨ Generate Random Student Data ✨
</div>

```python
# Generate data
# Using np.random.choice to pick strings, and np.random.randint for ages between 18 and 25
random_names = np.random.choice(first_names,n_student).tolist()
random_cities = np.random.choice(cities, n_student).tolist()
random_ages = np.random.randint(18, 26, n_student).tolist()
student_data = list(zip(random_names, random_cities, random_ages))
```
