    🌐 Process Of Web Scraping 🌐
    
</div>

- Sending HTTP request: send to target web server to retrieve desire web page
 - Retrieving web page content: Server responds with the web page content, usually HTML format
 - parsing the HTML : HTML content is parsed to locate specific data elements using libraries or tools designed for this purpose [BeautifulSoup,Scrapy]
 - Extracting data :relevant data is extracted from the parsed HTML structure based on predefined rules or criteria
 - Storing data : This says extracted data can be stored in various fromat such as csv, json, databases, for further analysis
 - Handling Challenges : Web Scrapers may need t address challenges like
    - Anti-scraping measures
    - dynamic content loading (javascript)
    - legal considerations

```python
import requests
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

url = "https://www.goodreads.com/quotes"
response = requests.get(url,headers=headers)
response.status_code
```

```python
from bs4 import  BeautifulSoup
import pandas as pd
```

```python
code = BeautifulSoup(html_code, 'html.parser')
code
```

```python
print(code.prettify())
```

<div style="
    background: #FADADD;
    color: #6D597A;
    padding: 12px;
    border-radius: 10px;
    text-align: center;

    🌐 Step 4 : Extracting Data 🌐
    
</div>

```python
quotes = code.find_all('div',class_= 'quote')
quotes
```

```python
quote_data = []

for q in quotes:
    details_el = q.find('a', class_='quoteDetails')
    text_el = q.find('div', class_='quoteText')
    title_el = q.find('span', class_='authorOrTitle')
    footer_el = q.find('a', class_='right')
    img_el = q.find('img')
    details = details_el.get_text(strip=True) if details_el else ""
    text = text_el.get_text(strip=True) if text_el else ""
    title = title_el.get_text(strip=True) if title_el else ""
    footer_details = footer_el.get_text(strip=True) if footer_el else ""
    image_url = img_el['src'] if img_el and img_el.has_attr('src') else ""
    quote_data.append({
        'details': details,
        'text': text,
        'title': title,
        'footer': footer_details,
        'image_url' : image_url
    })
quote_data
```

```python
df = pd.DataFrame(quote_data)
df
```
